```python
# Install necessary packages
!pip install pandas numpy nltk sentence-transformers faiss-cpu transformers
!pip install --upgrade transformers
!pip install gradio pyngrok

# Imports
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sentence_transformers import SentenceTransformer
import faiss
from transformers import AutoTokenizer, AutoModelForCausalLM
from transformers import T5Tokenizer, T5ForConditionalGeneration
import torch
import gradio as gr
from pyngrok import ngrok
import json
```

```python


# Load data
corpus_df = pd.read_json('/content/corpus.json')
train_df = pd.read_json('/content/train.json')

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Preprocess text function
def preprocess_text(text):
    # Lowercase the text
    text = text.lower()
    # Optionally, remove unwanted characters
    text = text.replace('\n', ' ').strip()
    return text

# Preprocess corpus
corpus_df['preprocessed_body'] = corpus_df['body'].apply(preprocess_text)

# Load embedding model
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings
corpus_embeddings = embedding_model.encode(
    corpus_df['preprocessed_body'].tolist(),
    show_progress_bar=True
)
corpus_embeddings = np.array(corpus_embeddings).astype('float32')

# Build FAISS index
embedding_dimension = corpus_embeddings.shape[1]
index = faiss.IndexFlatL2(embedding_dimension)
index.add(corpus_embeddings)

# Load FLAN-T5 model and tokenizer
model_name = 'google/flan-t5-base'
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Move model to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

# Function to extract key sentences
def extract_key_sentences(text, num_sentences=3):
    sentences = nltk.sent_tokenize(text)
    key_sentences = sentences[:num_sentences]
    return ' '.join(key_sentences)

# Function to extract facts
def extract_fact(text, query, num_sentences=2):
    # Tokenize text into sentences
    sentences = nltk.sent_tokenize(text)
    # Compute embeddings for sentences
    sentence_embeddings = embedding_model.encode(sentences)
    # Compute embedding for query
    query_embedding = embedding_model.encode([query])[0]
    # Compute similarities
    similarities = np.dot(sentence_embeddings, query_embedding)
    # Get indices of top sentences
    top_indices = similarities.argsort()[-num_sentences:][::-1]
    # Extract top sentences
    fact = ' '.join([sentences[i] for i in top_indices])
    return fact

# Function to process query
def process_query(query):
    # Preprocess the query
    preprocessed_query = preprocess_text(query)
    query_embedding = embedding_model.encode([preprocessed_query])
    query_embedding = np.array(query_embedding).astype('float32')

    # Search in FAISS index
    k = 4
    distances, indices = index.search(query_embedding, k)
    retrieved_indices = indices[0]
    retrieved_docs = corpus_df.iloc[retrieved_indices]

    # Build summarized context
    summarized_context = ""
    for idx, row in retrieved_docs.iterrows():
        key_sentences = extract_key_sentences(row['body'])
        summarized_context += f"Title: {row['title']}\n"
        summarized_context += f"Key Points: {key_sentences}\n\n"

    # Prepare the prompt
    prompt = f"""
    Answer the following question using the provided context.

    Question: {query}

    Context:
    {summarized_context}

    Answer:
    """

    # Tokenize the input
    input_ids = tokenizer.encode(prompt, return_tensors='pt', truncation=True, max_length=512).to(device)

    # Generate the answer
    outputs = model.generate(
        input_ids,
        max_length=150,
        num_beams=5,
        early_stopping=True
    )

    # Decode the generated answer
    generated_answer = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract evidence
    evidence_list = []
    for idx, row in retrieved_docs.iterrows():
        fact = extract_fact(row['body'], query)
        evidence = {
            "title": row['title'],
            "author": row['author'],
            "url": row['url'],
            "source": row['source'],
            "category": row['category'],
            "published_at": row['published_at'].strftime('%Y-%m-%d %H:%M:%S') if pd.notnull(row['published_at']) else None,
            "fact": fact
        }
        evidence_list.append(evidence)

    output = {
        "query": query,
        "answer": generated_answer,
        "question_type": "inference_query",
        "evidence_list": evidence_list
    }

    return output

# Function to format output
def format_output(output):
    formatted_output = f"Answer:\n{output['answer']}\n\n"
    formatted_output += "Evidence:\n"
    for evidence in output['evidence_list']:
        formatted_output += f"Title: {evidence['title']}\n"
        formatted_output += f"Author: {evidence['author']}\n"
        formatted_output += f"URL: {evidence['url']}\n"
        formatted_output += f"Source: {evidence['source']}\n"
        formatted_output += f"Category: {evidence['category']}\n"
        formatted_output += f"Published at: {evidence['published_at']}\n"
        formatted_output += f"Fact: {evidence['fact']}\n\n"
    return formatted_output

# Gradio function
def gradio_process_query(query):
    output = process_query(query)
    formatted_output = format_output(output)
    return formatted_output

# Create Gradio interface
iface = gr.Interface(
    fn=gradio_process_query,
    inputs=gr.Textbox(lines=2, placeholder="Enter your query here..."),
    outputs="text",
    title="Question Answering with Evidence",
    description="Enter a question and get an answer along with evidence from the corpus."
)

# Set up ngrok
from pyngrok import ngrok
# Kill any existing tunnels
ngrok.kill()

# Set your ngrok authentication token
!ngrok config add-authtoken 2nKvobAogGY1eWBnw3DBIOKBIKn_6HSLYUkA2iSfreYyTN3XY

# Start a new tunnel
public_url = ngrok.connect(7860)
print(f"Public URL: {public_url}")

# Launch the Gradio interface
iface.launch()
```
